```python
sns.countplot(x = 'Category', hue = 'Region', data = df, palette = 'twilight')
plt.xticks(rotation = 45, fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Category', fontsize = 15)
plt.ylabel('', fontsize = 20)
plt.legend(loc = 5, fontsize = 20);
```

```python
df["Ship Mode"].unique()
```

```python
sns.countplot(x = 'Ship Mode', hue = 'Region', data = df, palette = 'twilight_shifted')
plt.xticks(rotation=45,fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Ship modes', fontsize = 25)
plt.ylabel('', fontsize = 15)
plt.legend(loc = 5, fontsize = 15);
```

```python
plt.figure(figsize = (20, 9))
sns.countplot(x = df['State'],palette = 'twilight_shifted', order = (df['State'].value_counts().head(50)).index)
plt.xticks(rotation = 50, ha = 'right', fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('State', fontsize = 15)
plt.ylabel('Quantity ordered', fontsize = 10);
```

```python

plt.figure(figsize = (15, 7))
sns.countplot(x = df['City'],palette = 'twilight_shifted', order = (df['City'].value_counts().head(50)).index)
plt.xticks(rotation = 55, ha = 'right', fontsize = 10)
plt.yticks(fontsize = 15)
plt.xlabel('City', fontsize = 15)
plt.ylabel('Quantity ordered ', fontsize = 10);
```
